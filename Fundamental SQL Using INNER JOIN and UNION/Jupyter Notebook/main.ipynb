```python
import mysql.connector as db
import pandas as pd
from IPython.display import display

connection = db.connect(
                        host="localhost",
                        database="dqlab_academy",
                        user="root",
                        password="MySQL"
                        )

sql_inner_join = """select
                    tr_penjualan.kode_transaksi,
                    tr_penjualan.kode_pelanggan,
                    tr_penjualan.kode_produk,
                    ms_product.nama_produk,
                    tr_penjualan.qty,
                    ms_product.harga*tr_penjualan.qty total
                    from
                    tr_penjualan
                    inner join ms_product
                    on
                    tr_penjualan.kode_produk = ms_product.kode_produk
                    """

dataset = pd.read_sql(sql_inner_join, connection)

if connection.is_connected():
    connection.close()
    print("MySQL Connection is Close")

display(dataset)
```
