```python
import mysql.connector as db
import pandas as pd
from IPython.display import display

connection = db.connect(
                        host="localhost",
                        database="dqlab_academy",
                        user="root",
                        password="MySQL"
                        )
sql_ms_item_kategori = "select * from ms_item_kategori"
sql_ms_item_warna = "select * from ms_item_warna"

ms_item_kategori = "select * from ms_item_kategori"
ms_item_warna = "select * from ms_item_warna"

df_ms_item_kategori = pd.read_sql(ms_item_kategori, connection)
df_ms_item_warna = pd.read_sql(ms_item_warna, connection)

merge_inner_join = pd.merge(df_ms_item_kategori, df_ms_item_warna, how="inner", left_on="nama_item", right_on="nama_barang")

if connection.is_connected():
    connection.close()
    print("MySQL Connection is Close")
 
display(merge_inner_join)
```
